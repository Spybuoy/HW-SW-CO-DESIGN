```python
import random #used for generating the random matrices
import numpy as np #used for checking if my answer is right
import time #used to calculate the time of execution
```

```python
def matrix_multiplication(a: list, b: list):
    """ Performs O(N^3) matrix multiplication """
    # TODO: implement matrix multiplication
    result = []
    l = len(a)
    for i in range(l):
      temp_row = []
      for j in range(l):
          temp_element = 0
          for k in range(l):
              temp_element += a[i][k] * b[k][j]
          temp_row.append(temp_element)
      result.append(temp_row)
    return result

# TODO: Generate random matrices using float
N = 3
A_3 = [[round(random.random(),2) for i in range(N)] for j in range(N)]
B_3 = [[round(random.random(),2) for i in range(N)] for j in range(N)]
start_time = time.perf_counter ()
C_3 = matrix_multiplication(A_3, B_3)
end_time = time.perf_counter ()
t = end_time - start_time
print('naive_runtime_'+str(N)+': '+str(t)+'s')

N = 4
A_4 = [[round(random.random(),2) for i in range(N)] for j in range(N)]
B_4 = [[round(random.random(),2) for i in range(N)] for j in range(N)]
start_time = time.perf_counter ()
C_4 = matrix_multiplication(A_4, B_4)
end_time = time.perf_counter ()
t = end_time - start_time
print('naive_runtime_'+str(N)+': '+str(t)+'s')

A_33 = np.matrix(A_3)
B_33 = np.matrix(B_3)

C_33 = A_33@B_33

A_44 = np.matrix(A_4)
B_44 = np.matrix(B_4)

C_44 = A_44@B_44

def rounding(a : list):
  rounded_list = [round(num,2) for num in a]
  return rounded_list
def rounding2(a : list):
  op = []
  for each in a:
    temp = rounding(each)
    op.append(temp)
  return op

# TODO: Verify that the result is correct for both a 3x3 and 4x4 matrix
C_3 = rounding2(C_3)
C_33 = C_33.round(2)
assert (C_3 == C_33).all()

C_4 = rounding2(C_4)
C_44 = C_44.round(2)
assert (C_4 == C_44).all()
```
